```python
def read_data_11(file_name):
    with open(file_name) as my_file:
        input = my_file.read()
    input_strings = input.splitlines()
    devices = {}
    for line in input_strings:
        parts = line.split(": ")
        devices[parts[0]] = parts[1].split(" ")
    
    return devices
```

```python
devices_test = read_data_11("../data/input11_test.txt")
devices_test2 = read_data_11("../data/input11_test2.txt")
devices_puzzle = read_data_11("../data/input11_puzzle.txt")
devices_test
```

## Part 1

```python
def find_paths(devices_dict, at_device, out_device, current_path, found_paths):
    current_path = current_path.copy() + [at_device]
    if at_device == out_device:
        # print("found solution")
        found_paths.append(current_path)
    else:
        if not at_device == "out":
            for device in devices_dict[at_device]:
                # print(f"{at_device} -> {device}")
                if not device in current_path:
                    found_paths = find_paths(devices_dict, device, out_device, current_path, found_paths)
    return found_paths
```

```python
found_paths = find_paths(devices_test, "you", "out", [], [])
for id, path in enumerate(found_paths):
    print(id, path)
len(found_paths)
```

## Part 2

```python
def find_paths_w_fft_and_dac(devices_dict, at_device, out_device, current_path, found_paths):
    current_path = current_path.copy() + [at_device]
    if at_device == out_device:
        # print("found solution")
        if "fft" in current_path and "dac" in current_path:
            found_paths.append(current_path)
    else:
        if not at_device == "out":
            for device in devices_dict[at_device]:
                # print(f"{at_device} -> {device}")
                if not device in current_path:
                    found_paths = find_paths_w_fft_and_dac(devices_dict, device, out_device, current_path, found_paths)
    return found_paths
```

```python
def count_paths_w_fft_and_dac(devices_dict, at_device, out_device, current_path, counter):
    current_path = current_path.copy() + [at_device]
    if at_device == out_device:
        # print("found solution")
        if "fft" in current_path and "dac" in current_path:
            counter += 1
    else:
        if not at_device == "out":
            for device in devices_dict[at_device]:
                # print(f"{at_device} -> {device}")
                if not device in current_path:
                    counter = count_paths_w_fft_and_dac(devices_dict, device, out_device, current_path, counter)
    return counter
```
